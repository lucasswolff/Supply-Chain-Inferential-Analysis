# Inferential Statistical Analysis on Supply Chain Data

This notebook details how Inferential Statistical Analysis can be used on a Supply Chain dataset.
There're several scenarios to implament this testing in Supply Chain, such as:

• Best strategy to reduce inventory turnover

• Compare two shipping methods to find the most cost effective

• Compare two quality inspection methods

In this notebook, we will compare the profit per order between two regions, to check each is more profitable, and if the sample data has enough statistical evidence to make a statement about the population

## Inferential Statistical Analysis

Now it's time to finally answer the question we've asked:
Imagine that a logists manager would like to know if the company is getting more profit from shippimento to Latin America or Europe. By doing that, they can take new strategies on the company, maybe trying to make one of these markets more profitable, or focusing only in the most profitable.

So you, as an analyst, gathered some sample data from past orders, in order to analyse that. Since this is a sample data, we can not only compare the means. We need to check if the mean from both reagions is statistically significant. Meaning to check if that is probably a real difference in the population (all orders) or that difference only occured by random chance in this sample.

In summary, the question is: **in 2017, for the orders shipped from the US, the mean profit by order in Europe is bigger than in Latim America?**  

So this a problem of Inferential Statistical Analysis comparing proportion of two independent samples.
Our hypotheses are:

• **Null Hypothesis (H0):** The mean profit from shipments to Europe is equal to or less than the mean profit from shipments to Latin America.

• **Alternative Hypothesis (H1):** The mean profit from shipments to Europe is greater than the mean profit from shipments to Latin America.

We'd like to test that with a 0.05 significance level

We begging by filtering or data for only these regions

```python
df_US_2017 = pd.DataFrame(gdf_US_2017)
df_US_2017 = df_US_2017[['Market', 'Benefit per order']]
```

```python
df_US_2017.head()
```

```python
df_US_2017_eur_latam = df_US_2017[(df_US_2017['Market'] == 'LATAM') | (df_US_2017['Market'] == 'Europe')]
```

Now let's analyse the data

```python
df_US_2017_eur_latam_agg = df_US_2017_eur_latam.groupby('Market')['Benefit per order'].agg([lambda x: np.mean(x), 
                                                                                            lambda x: np.var(x),
                                                                                            'size'])
df_US_2017_eur_latam_agg.columns = ["Mean", "Variance","size"]
df_US_2017_eur_latam_agg
```

To perform the statistical test, we need to follow some assumptions about the data:

• The observations in each group must be **independent** of each other. 

• The data in each group should follow an approximately **normal distribution**. If not, each sample should be **larger enough (>30)** to rely on the Central Limit Theorem)

• The **variances** of the two groups should be **approximately equal.**

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

#plot for LATAM
sns.histplot(data=df_US_2017_eur_latam[df_US_2017_eur_latam['Market'] == 'LATAM'], x='Benefit per order', bins=50, ax=axes[0], color='blue')
axes[0].set_title('LATAM')

#plot for Europe
sns.histplot(data=df_US_2017_eur_latam[df_US_2017_eur_latam['Market'] == 'Europe'], x='Benefit per order', bins=50, ax=axes[1], color='green')
axes[1].set_title('Europe')

fig.suptitle('Histograms of Profit per Order for LATAM and Europe Markets')

plt.show()
```

The data is left skwed, especially in LATAM. Luckly, our sample is large enough, so we can rely on the CTL.
But as you can see, the variances are not similar. Therefore, we can use the Welch's t-test, which is designed for unequal population variances, but still holding on the normality assumption. 

For more informations about Welch's t-test: https://www.statisticshowto.com/welchs-test-for-unequal-variances/

```python
df_US_2017_eur = df_US_2017_eur_latam[df_US_2017_eur_latam['Market'] == 'Europe']['Benefit per order']
df_US_2017_latam = df_US_2017_eur_latam[df_US_2017_eur_latam['Market'] == 'LATAM']['Benefit per order']
```

```python
t_statistic, p_value = stats.ttest_ind(df_US_2017_eur, df_US_2017_latam, equal_var=False, alternative = 'greater')
print('T statistic: ' + str(t_statistic))
print('P value: ' + str(p_value))
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

**Conclusion:** our p value is less than our alpha. This result provides sufficient evidence to reject the null hypothesis. Therefore, based on the sample data and the statistical analysis performed, we can state that the average profit per order in Europe is greater than that in Latin America.
